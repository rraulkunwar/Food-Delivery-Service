```python
hot_items = ['Pasta', 'Pizza', 'Lasagna', 'Risotto', 'Burrito', 'Tacos', 'Enchiladas', 'Chiles Rellenos', 'Hot pot', 'Noodles', 'Dumpling', 'Peking Duck', 'Kung Pao Chicken', 'Hot Soup', 'Pad Thai', 'Green Curry', 'Butter Chicken', 'Palak Paneer', 'Biryani', 'Tandoori Chicken', 'Masala Chai', 'Sushi', 'Ramen', 'Tempura', 'Udon', 'Moussaka', 'Souvlaki', 'Burger', 'Fries', 'Steak', 'BBQ Ribs', 'Clam Chowder']
cold_items = ['Tiramisu', 'Gelato', 'Prosecco', 'Limoncello', 'Margerita', 'Horchata', 'Michelada', 'Tequila', 'Bubble Tea', 'Jasmine Tea', 'Soy Milk', 'Spring Roll', 'Thai Iced Tea', 'Coconut Water', 'Lassi', 'Cake', 'Sake', 'Matcha Tea', 'Ramune Soda', 'Ice-cream', 'Ouzo', 'Greek Coffee', 'Mythos Beer', 'Coke', 'Root Beer', 'Milkshake', 'Iced Coffee']

Temperature = ['Room','Hot','Cold']
hot_food_probabilities = [0.33,0.47,0.20]
cold_food_probabilities = [0.15,0.28,0.57]

# Function to assign delivered temperature (modified)
def assign_delivered_temperature(row):
    if row['Item'] in hot_items:
        return random.choices(Temperature, hot_food_probabilities)[0]
    elif row['Item'] in cold_items:
        return random.choices(Temperature, cold_food_probabilities)[0]
    else:  
        return random.choice(Temperature) 

# Create the new column
df_test["Delivered_temperature"] = df_test.apply(assign_delivered_temperature, axis=1)

df_test

```

```python
# Define a function to determine the taste based on item and temperature
def determine_taste(row):
    hot_items = ['Pasta', 'Pizza', 'Lasagna', 'Risotto', 'Burrito', 'Tacos', 'Enchiladas', 'Chiles Rellenos', 'Hot pot', 'Noodles', 'Dumpling', 'Peking Duck', 'Kung Pao Chicken', 'Hot Soup', 'Pad Thai', 'Green Curry', 'Butter Chicken', 'Palak Paneer', 'Biryani', 'Tandoori Chicken', 'Masala Chai', 'Sushi', 'Ramen', 'Tempura', 'Udon', 'Moussaka', 'Souvlaki', 'Burger', 'Fries', 'Steak', 'BBQ Ribs', 'Clam Chowder']
    cold_items = ['Tiramisu', 'Gelato', 'Prosecco', 'Limoncello', 'Margerita', 'Horchata', 'Michelada', 'Tequila', 'Bubble Tea', 'Jasmine Tea', 'Soy Milk', 'Spring Roll', 'Thai Iced Tea', 'Coconut Water', 'Lassi', 'Cake', 'Sake', 'Matcha Tea', 'Ramune Soda', 'Ice-cream', 'Ouzo', 'Greek Coffee', 'Mythos Beer', 'Coke', 'Root Beer', 'Milkshake', 'Iced Coffee']
    
    if row['Delivered_temperature'] == 'Cold' and row['Item'] in hot_items:
        return 'Low'
    elif row['Delivered_temperature'] == 'room' and row['Item'] in hot_items:
        return 'Moderate'
    elif row['Delivered_temperature'] == 'Room' and row['Item'] in cold_items:
        return 'Low'
    elif row['Delivered_temperature'] == 'Hot' and row['Item'] in cold_items:
        return 'Low'
    
    else:
        return 'High'

# Apply the function to the DataFrame
df['Freshness'] = df.apply(determine_taste, axis=1)

df
```

```python
# Define a function to determine the taste based on freshness
def determine_good_taste(row):
    if row['Freshness'] == 'Low':
        return 'Bad'
    elif row['Freshness'] == 'Moderate':
        return 'Average'
    else:
        return 'Good'

# Apply the function to the DataFrame
df['Food_quality'] = df.apply(determine_good_taste, axis=1)

df

```

```python
# Define your order count array
order_count = [903, 851, 829, 801, 779, 761, 739, 671, 859, 881, 953, 973]

# Random date generation function
def random_date(start, end):
    return start + timedelta(
        seconds=random.randint(0, int((end - start).total_seconds())))

# Define the start and end dates
start = datetime(2023, 1, 1)
end = datetime(2024, 1, 1)

# Generate the DataFrame
dates = []
for month, count in enumerate(order_count):  # Iterate over months and counts
    start_month = start.replace(month=month + 1)  
    if month == 11:  # Adjust for December
        end_month = end
    else:
        end_month = start.replace(month=month + 2)
    for _ in range(count):   # Generate the specified number of rows for each month 
        dates.append(random_date(start_month, end_month))

# Create the new DataFrame
new_df = pd.DataFrame({
    'Customer Placed Order time': dates
})

# Assuming you have an existing DataFrame called 'df' 
# Concatenate the new DataFrame with the existing one
df = pd.concat([df, new_df], axis=1)

# Display the combined DataFrame
df
```

```python
def order_accepted_time(row):
    # Get the placed time
    placed_time = row['Customer Placed Order time']

    # Add a random delay of 2 to 5 minutes
    accepted_time = placed_time + timedelta(minutes=random.randint(2, 5), seconds=random.randint(0, 59))

    return accepted_time

# Apply the function to the DataFrame
df['Order Accepted by Restaurant time'] = df.apply(order_accepted_time, axis=1)

df
```

```python
def Driver_at_Restaurant_Datetime(row):
    # Get the order accepted time
    order_accepted_time = row['Order Accepted by Restaurant time']

    # Add a random delay of 7 to 10 minutes
    driver_time = order_accepted_time + timedelta(minutes=random.randint(7, 10), seconds=random.randint(0, 59))

    return driver_time

# Apply the function to the DataFrame
df['Driver at Restaurant Datetime'] = df.apply(Driver_at_Restaurant_Datetime, axis=1)

df
```

```python
def Delivered_to_Consumer_Datetime(row):
    # Get the driver at restaurant time
    driver_at_restaurant_time = row['Driver at Restaurant Datetime']

    # Add a random delay of 10 to 15 minutes
    deliver_time = driver_at_restaurant_time + timedelta(minutes=random.randint(10, 15), seconds=random.randint(0, 59))

    return deliver_time

# Apply the function to the DataFrame
df['Delivered to Consumer Datetime'] = df.apply(Delivered_to_Consumer_Datetime, axis=1)

df
```

```python
def calculate_wait_time(row):
    # Get the order time and delivery time
    order_time = row['Customer Placed Order time']
    delivery_time = row['Delivered to Consumer Datetime']

    # Calculate the difference in minutes
    wait_time_in_minutes = int((delivery_time - order_time).total_seconds() // 60)

    return wait_time_in_minutes

# Apply the function to the DataFrame
df["Wait Time in minutes"] = df.apply(calculate_wait_time, axis=1)

df
```

```python
df['Order Quantity'] = [random.randint(1, 5) for _ in range(num_records)]
# For 'Premium' and 'Monthly Premium' subscriptions, set delivery fee to 0
df.loc[df['Subscription'].isin(['Premium', 'Monthly Premium']), 'delivery fee'] = 0

# For 'Basic' and 'Monthly Basic' subscriptions, set delivery fee to one of the options: 0.99, 1.99, 2.99
basic_options = [0.99, 1.99, 2.99]
basic_indices = df[df['Subscription'].isin(['Basic', 'Monthly Basic'])].index
for i in basic_indices:
    df.loc[i, 'delivery fee'] = random.choice(basic_options)

# For 'Free' subscriptions, set delivery fee to one of the options: 2.99, 3.99, 4.99, 5.99
free_options = [3.99, 4.99, 5.99]
free_probs = [0.5,0.3,0.2]
free_indices = df[df['Subscription'] == 'Free'].index
for i in free_indices:
    df.loc[i, 'delivery fee'] = random.choices(free_options,free_probs)[0]

df
```

```python
df['Total Price'] = df['Meal Price'] * df['Order Quantity']
df['Service Fee'] = df['Total Price']*0.11
df['Tax'] = df['Total Price'] *0.145
df['Final Price'] = df['Total Price'] + df['Service Fee'] + df['Tax'] + df['Delivery fee']
df['Number of Claims'] = [random.randint(1,20) for _ in range(len(df))]
df['Complaints Resolved'] = [random.randint(int(claim/1.5), claim) for claim in df['Number of Claims']]
df['Number of Open Complaints'] = df['Number of Claims'] - df['Complaints Resolved']
df['Amount of Refund for resolved complaint'] = df['Meal Price'] * df['Complaints Resolved'] + (df['Meal Price'] * df['Complaints Resolved'])*0.145
df.loc[df['Complaints Resolved'] == 0, 'Amount of Refund for resolved complaint'] = 0
df
```
